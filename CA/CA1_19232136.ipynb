```python
#Q1

from pprint import pprint
import requests
import json
import pandas as pd
import csv
import sys

def airport_data():
    try:
        with open ('/Users/omkartawade/Omkar/DAP/Database-and-Analytics-Programming/CA/DublinAirport090220.json') as f:
            data = json.load(f)
    except IOError: 
        print('cannot open json file')        #If file not found then this exception will be called
    return data

def json2csv(data):
    airportdata = data
    ap_data = open('/Users/omkartawade/Omkar/DAP/Database-and-Analytics-Programming/CA/airportdata.csv','w')  #csv file created
    csvwriter = csv.writer(ap_data)       
    count = 0
    for x in airportdata:
        if count == 0:
            header = x.keys()
            csvwriter.writerow(header)    #column name is copied to csv file
            count = count+1
        csvwriter.writerow(x.values())    #value are filled in csv file
    ap_data.close()
    
def print_record(data):
    airportdata = data
    #Convert List to Dataframe
    df = pd.DataFrame(airportdata, columns=['name', 'temperature','symbol','weatherDescription','text','windSpeed','windGust','cardinalWindDirection','windDirection','humidity','rainfall','pressure','dayName','date','reportTime'])
    #Accesing Dataframe rowwise
    for index,row in df.iterrows():
        print("\nRecord "+str(index+1)+":")
        print("temperature: {}".format(row['temperature']))
        print("weatherDescription: {}".format(row['weatherDescription']))
        print("reportTime: {}".format(row['reportTime']))
    
def main():
    airportdata=[]
    airportdata = airport_data()
    print_record(airportdata)
    json2csv(airportdata)    
    
if __name__=="__main__":
    main()

```

```python
#Q2

import numpy as np
array = np.arange(2,401,2)  #Created a 1d array
print (array)

array2d = np.reshape(array, (50,4))   #Reshaped 1d array to 2d array
print(array2d[0])

array4c = array2d[...,3:]       #Sliced first 3 columns of array2d and 4th column of array 2d is copied into array4c
print(array4c)

array2r = np.empty((0,4),int)  #Created an empty 2d array

for x in range(0,50,2):
    array2r = np.append(array2r,array2d[x])   #Copied alternate row of array2d into array 2r
print(array2r)
```
