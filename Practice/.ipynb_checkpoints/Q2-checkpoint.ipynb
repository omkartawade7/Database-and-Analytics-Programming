```python
import numpy as np
```

```python
#Numpy array filled with 800 numbers ranges from 1 to 100 in 200rows * 4columns
data = np.random.randint(100, size=(200,4))    
```

```python
#Slicing array into 4 different array

data1 = data[0:50]
data2 = data[50:100]
data3 = data[100:150]
data4 = data[150:200]

print(data[0])
print(data1[0])

print(data[51])
print(data2[0])

print(data[101])
print(data3[0])

print(data[151])
print(data4[0])
```

```python
#Resshaping sliced arrays into different shapes

data1new = data1.reshape(20,10)
data2new = data2.reshape(2,100)
data3new = data3.reshape(100,2)

print(data1new[0])
print(data1new[1])
print(data2new[0])
print(data3new[0])
```

```python
#Spliting array horizontally


np.vsplit(data1new,2)
np.hsplit(data3new,2)

```
