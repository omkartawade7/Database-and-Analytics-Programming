```python
import numpy as np
```

```python
#Numpy array filled with 800 numbers ranges from 1 to 100 in 200rows * 4columns
data = np.random.randint(100, size=(200,4))    
```

```python
#Slicing array into 4 different array

data1 = data[0:49]
data2 = data[50:99]
data3 = data[100:149]
data4 = data[150:199]

print(data1[0])
print(data1[0])

print(data[50])
print(data2[0])

print(data[100])
print(data3[0])

print(data[150])
print(data4[0])
```

```python
#Resshaping sliced arrays into different shapes

data1new = data1.reshape(98,2)
data2new = data2.reshape(49,4)
data3new = data3.reshape(14,14)

print(data1[0])
print(data2[0])
print(data3[0])
```

```python
#Spliting array horizontally

np.hsplit(data1,2)
np.hsplit(data2,4)

```
